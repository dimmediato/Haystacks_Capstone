```python
#Import required libraries
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np

'''This is a notebook that will be used to conduct statistical comparisons between the models used in the other notebooks. In those other notebooks,
the model scores should have been converted to pickle files, and can then be imported here. We already know which scores are superior to the others, but
we don't know if it's significant.'''

#Import pickle files
baseline_scores = pd.read_pickle("./Scores/baseline_scores.pkl")
baseline_MLR_scores = pd.read_pickle("./Scores/baseline_MLR_scores.pkl")
baseline_ridge_scores = pd.read_pickle("./Scores/baseline_ridge_scores.pkl")
distance_scores = pd.read_pickle("./Scores/distance_scores.pkl")
distance_MLR_scores = pd.read_pickle("./Scores/distance_MLR_scores.pkl")
distance_ridge_scores = pd.read_pickle("./Scores/distance_ridge_scores.pkl")
density_one_scores = pd.read_pickle("./Scores/density_one_scores.pkl")
density_three_scores = pd.read_pickle("./Scores/density_three_scores.pkl")
density_five_scores = pd.read_pickle("./Scores/density_five_scores.pkl")
density_MLR_one_scores = pd.read_pickle("./Scores/density_MLR_one_scores.pkl")
density_ridge_one_scores = pd.read_pickle("./Scores/density_ridge_one_scores.pkl")
density_MLR_three_scores = pd.read_pickle("./Scores/density_MLR_three_scores.pkl")
density_ridge_three_scores = pd.read_pickle("./Scores/density_ridge_three_scores.pkl")
density_MLR_five_scorespd = pd.read_pickle("./Scores/density_MLR_five_scores.pkl")
density_ridge_five_scores = pd.read_pickle("./Scores/density_ridge_five_scores.pkl")
```

```python
# Perform paired t-tests for all columns (metrics) at once and compute Cohen's d
def paired_t_tests_with_cohens_d(model1_df, model2_df):
    # Dictionary to store t-statistics, p-values, and Cohen's d
    t_test_results = {}
    
    # Iterate over the columns (metrics)
    for column in model1_df.columns:
        # Compute differences
        differences = model1_df[column] - model2_df[column]
        
        # Perform paired t-test for each metric
        t_stat, p_val = stats.ttest_rel(model1_df[column], model2_df[column])
        
        # Compute Cohen's d
        mean_diff = np.mean(differences)
        std_diff = np.std(differences, ddof=1)  # Use ddof=1 for sample standard deviation
        cohens_d = mean_diff / std_diff if std_diff != 0 else np.nan
        
        # Store results
        t_test_results[column] = {'t-stat': t_stat, 'p-value': p_val, "Cohen's d": cohens_d}
    
    # Convert results to a DataFrame for easy viewing
    results_df = pd.DataFrame(t_test_results).T
    
    # Round the values to the nearest 5th decimal place
    results_df = results_df.round(5)
    
    return results_df

# Example usage
t_test_results_df = paired_t_tests_with_cohens_d(density_three_scores, density_five_scores)

# Display the results
print("Significance Between Distance and the Baseline Scores")
t_test_results_df
```
